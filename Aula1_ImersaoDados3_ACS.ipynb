####  1.1 Importar tabela 'zipada' do github da alura utilizando biblioteca pandas


Na linnha 6 ordenamos que o python leia a tabela csv (read_csv), separada por vírgulas, e indicamos que ela está compactada .zip. É comum encontrarmos em documentações o nome atribuído à variável da tabela como 'df', isso vem de DataFrame. Aqui iremos nomeá-la como 'dados'.

```python
# é convencional dar o "apelido" pd para a biblioteca pandas
import pandas as pd

url_dados= 'https://github.com/alura-cursos/imersaodados3/blob/main/dados/dados_experimentos.zip?raw=true'

dados = pd.read_csv(url_dados, compression = 'zip')
dados
```

#### 1.3 Frequência Relativa


*   .value_counts(normalize = True)

> Lembre que a soma de todas as frequências relativas deve ser 1.

> Pois Frequência Relativa é uma proporção obtida a partir da divisão da Frequência Absoluta, a ocorrência de dado valor, pelo número total de valores. 

> Exemplo: há duas categorias para 'dose' > D1 e D2. As frequências absolutas são D1 = 12147 e
D2 = 11667. A soma das frequências absolutas é = 23814. 

> A frequência relativa de D1 é = 12147/23814 = 0,510078 e a de D2 é =  11667/23814 = 0,489922.



```python
dados.dose.value_counts(normalize = True)
```

```python
dados.tempo.value_counts(normalize = True)
```

```python
dados.tratamento.value_counts(normalize = True)
```

#### 1.6 Desafios


######  Desafio 01: Indique por que a classe 'tratamento' é tão desbalanceada




O valor correspondente ao tratamento controle é de 7.84% amostras e ao tratamento com composto/droga é de 92,16% amostras. Experimentos científicos apresentam um valor de amostras controle relevantes ao experimento em , sendo assim não é necessário obter o mesmo número de amostras com algum tipo de tratamento em relação às amostras controle.


[Gráficos de pizza com Matplotlib](https://medium.com/horadecodar/gr%C3%A1ficos-de-pizza-com-matplotlib-piechart-33ea9760ad87)


```python
# gráfico de pizza(torta)
tipos_tratamento = dados.tratamento.value_counts()
labels = 'Droga', 'Controle'
desloca_controle = (0, 0.4)
cores = ['mediumaquamarine', 'plum']

plt.title('Tipos de tratamento')
plt.pie(tipos_tratamento, labels = labels, shadow = True, explode = desloca_controle, autopct = '%1.2f%%', colors = cores)

plt.show()
```

###### Desafio 05: Procurar na documentação o método query() do pandas


[.query()](https://pandas.pydata.org/docs/reference/api/pandas.DataFrame.query.html)

> O método .query() nos permite realizar consultas nas colunas de um DataFrame com uma expressão booleana.
