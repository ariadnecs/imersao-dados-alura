####  1.1 Importar tabela 'zipada' do github da alura utilizando biblioteca pandas

#### 1.3 Frequncia Relativa


*   .value_counts(normalize = True)

> Lembre que a soma de todas as frequncias relativas deve ser 1.

> Pois Frequncia Relativa  uma proporo obtida a partir da diviso da Frequncia Absoluta, a ocorrncia de dado valor, pelo nmero total de valores. 

> Exemplo: h duas categorias para 'dose' > D1 e D2. As frequncias absolutas so D1 = 12147 e
D2 = 11667. A soma das frequncias absolutas  = 23814. 

> A frequncia relativa de D1  = 12147/23814 = 0,510078 e a de D2  =  11667/23814 = 0,489922.



```python
dados.dose.value_counts(normalize = True)
```

```python
dados.tempo.value_counts(normalize = True)
```

```python
dados.tratamento.value_counts(normalize = True)
```

######  Desafio 01: Indique por que a classe 'tratamento'  to desbalanceada




O valor correspondente ao tratamento controle  de 7.84% amostras e ao tratamento com composto/droga  de 92,16% amostras. Experimentos cientficos apresentam um valor de amostras controle relevantes ao experimento em , sendo assim no  necessrio obter o mesmo nmero de amostras com algum tipo de tratamento em relao s amostras controle.


[Grficos de pizza com Matplotlib](https://medium.com/horadecodar/gr%C3%A1ficos-de-pizza-com-matplotlib-piechart-33ea9760ad87)


```python
# grfico de pizza(torta)
tipos_tratamento = dados.tratamento.value_counts()
labels = 'Droga', 'Controle'
desloca_controle = (0, 0.4)
cores = ['mediumaquamarine', 'plum']

plt.title('Tipos de tratamento')
plt.pie(tipos_tratamento, labels = labels, shadow = True, explode = desloca_controle, autopct = '%1.2f%%', colors = cores)

plt.show()
```
